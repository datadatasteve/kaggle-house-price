```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configure visualization settings
plt.style.use('ggplot')
sns.set_palette("husl")
%matplotlib inline

# Load the data
train_df = pd.read_csv('../train.csv')

print("Data loaded successfully!")
print(f"Dataset shape: {train_df.shape}")
```

```python
# Display the first few rows
print("First 5 rows of the dataset:")
train_df.head()
```

```python
# see all column names/data types
print("Column INformation:")
print(f"Total columns: {len(train_df.columns)}\n")
train_df.info()
```

```python
#stats summary of numerical columns
train_df.describe()
```

```python
#a sample of the data
#Check columns details
print("\nColumns with 'Year' in the name:")
year_cols = [col for col in train_df.columns if 'Year' in col or 'Yr' in col]
print(year_cols)

print("\nColumns with 'Sale' in the name:")
sale_cols = [col for col in train_df.columns if 'Sale' in col]
print(sale_cols)

print("\nA sample of these columns:")
train_df[year_cols + sale_cols].head(10)
```

```python
# Analysis - renovated vs not
train_df['WasRemodeled'] = train_df['YearRemodAdd'] != train_df['YearBuilt']

print("Renovation Breakdown:")
print(train_df['WasRemodeled'].value_counts())
print(f"\Percentage renovated: {train_df['WasRemodeled'].mean()*100:.1f}%")

#compare Sales Price
print("\nSale Price Comparison:")
print(f"Renovated homes - Average: ${train_df[train_df['WasRemodeled']==True]['SalePrice'].mean():,.0f}")
print(f"Non-renovated homes - Average: ${train_df[train_df['WasRemodeled']==False]['SalePrice'].mean():,.0f}")
```

```python
# Visualize the price difference
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Box plot comparing prices
train_df.boxplot(column='SalePrice', by='WasRemodeled', ax=axes[0])
axes[0].set_title('Sale Price: Renovated vs Not Renovated')
axes[0].set_xlabel('Was Remodeled')
axes[0].set_ylabel('Sale Price ($)')

# Look at age vs price
train_df['HouseAge'] = train_df['YrSold'] - train_df['YearBuilt']
axes[1].scatter(train_df['HouseAge'], train_df['SalePrice'], alpha=0.5)
axes[1].set_title('House Age vs Sale Price')
axes[1].set_xlabel('Age at Sale (years)')
axes[1].set_ylabel('Sale Price ($)')

plt.tight_layout()
plt.show()

print(f"\nAverage house age at sale: {train_df['HouseAge'].mean():.1f} years")
```

```python
# improved graph for renovated/not-renovated
fig, ax = plt.subplots(figsize=(10, 6))

train_df[train_df['WasRemodeled']==False]['SalePrice'].hist(bins=30, alpha=0.5, label='Not Renovated', ax=ax)
train_df[train_df['WasRemodeled']==True]['SalePrice'].hist(bins=30, alpha=0.5, label='Renovated', ax=ax)

ax.set_xlabel('Sale Price ($)')
ax.set_ylabel('Number of Houses')
ax.set_title('Sale Price Distribution: Renovated vs Not Renovated')
ax.legend()
plt.show()
```

```python
# Better colors - blue and orange
fig, ax = plt.subplots(figsize=(10, 6))

train_df[train_df['WasRemodeled']==False]['SalePrice'].hist(bins=30, alpha=0.6, label='Not Renovated', color='blue', ax=ax)
train_df[train_df['WasRemodeled']==True]['SalePrice'].hist(bins=30, alpha=0.6, label='Renovated', color='orange', ax=ax)

ax.set_xlabel('Sale Price ($)')
ax.set_ylabel('Number of Houses')
ax.set_title('Sale Price Distribution: Renovated vs Not Renovated')
ax.legend()
plt.show()
```
